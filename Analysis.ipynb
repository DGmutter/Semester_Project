```python

elements = pd.read_csv("elements-by-episode.csv")

element_columns = elements.drop(columns=["EPISODE", "TITLE"])

element_counts = element_columns.sum()

top5 = element_counts.sort_values(ascending=False).head(5)

plt.bar(top5.index, top5.values)
plt.title("Top 5 Most Common Elements")
plt.xlabel("Element")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
 


elements = pd.read_csv("elements-by-episode.csv")


elements["SEASON"] = elements["EPISODE"].str.extract(r"S(\d+)").astype(int)


elements["TOTAL_TREES"] = elements["TREE"] + elements["TREES"]


trees_per_season = elements.groupby("SEASON")["TOTAL_TREES"].sum()


plt.bar(trees_per_season.index, trees_per_season.values)
plt.title("Total Trees Per Season")
plt.xlabel("Season")
plt.ylabel("Number of Trees")
plt.tight_layout()
plt.show()
```

```python
 


elements = pd.read_csv("elements-by-episode.csv")


all_titles = " ".join(elements["TITLE"])


all_titles = all_titles.lower()


all_titles = all_titles.replace(",", "").replace(".", "").replace("'", "")


words = all_titles.split()


word_counts = pd.Series(words).value_counts()


top5_words = word_counts.head(5)


plt.bar(top5_words.index, top5_words.values)
plt.title("Top 5 Most Common Words in Episode Titles")
plt.xlabel("Word")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
